REMOVING NULL VALUES FROM THE DATASET

```python
df1=df.dropna()
df1
```

LABELLING THE TARGET VALUES

```python
dict={"Basmati":0,"Arborio":1,"Jasmine":2,"Ipsala":3,"Karacadag":4}
df1["CLASS"]=df1["CLASS"].map(dict)
df1["CLASS"].value_counts()
```

DIVIDING THE DATASET INTO INDEPENDENT VALUES AND TARGET VALUES

```python
x=df1.iloc[0:,:-1] 
print(x)
y=df1["CLASS"]
print(x.shape)
print(y.shape)
print(y)
```

DATA SPLITTING FOR TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
##STANDARDIZING
sc.fit(x_train)
x_train_std=sc.transform(x_train)
x_test_std=sc.transform(x_test)
```

MODEL 1: LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
clf1=LogisticRegression()
clf1.fit(x_train,y_train)
y_test_pred=clf1.predict(x_test)
y_train_pred=clf1.predict(x_train)
```

```python
clf1.fit(x_train_std,y_train)
y_test_pred=clf1.predict(x_test_std)
y_train_pred=clf1.predict(x_train_std)
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print("Training Acc=",train_acc)
print("Testing Acc=",test_acc)
```

```python
cr1=classification_report(y_test,y_test_pred)
cr1
```

```python
cm1=confusion_matrix(y_test,y_test_pred)
print(cm1)
```

```python
sns.heatmap(cm1,annot=True,cbar=False)
```

```python
from sklearn.linear_model import Perceptron
clf2=LogisticRegression()
clf2.fit(x_train,y_train)
y_test_pred=clf2.predict(x_test)
y_train_pred=clf2.predict(x_train)
```
